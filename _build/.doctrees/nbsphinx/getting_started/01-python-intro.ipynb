### 1.2 Collections (Lists, Tuples, and Dictionaries)
**Lists** store a series of mutable items in a particular order. You access items using an index, or within a loop. They help by keeping related data within the same structure so the same operations can be performed on many values at the same time.

Here are a few ways to initialize lists in Python.

```python
list1 = [1.2, 1.24, 1.59, 5.43, 2.99] # comma separated items, with square brackets framing the outside
print(list1)
list2 = list((1,2,3,4,5)) # using Python's builtin list() function to create a list with the parameter values
print(list2)

# you can also combine the lists, or create nested lists as such:
combined = list1 + list2
print(f'Combined list {combined}')
nested = [list1, list2]
print(f'Nested list {nested}')
```

Items can also be added and removed from a list. To add an item, the `.append(<item>)` function can be used to add the item to the end of an existing list, or `.insert(<index>, <item>)` can add the item at a specified index. *Remember, index starts at 0 not 1!*

```python
vowels = ['a','i','o']
print(vowels)
# now let's add the vowel 'u' to the end of the list:
vowels.append('u')
print(vowels)
# let's add in the vowel 'e' between 'a' (index 0) and 'i' 
vowels.insert(1, 'e')
print(vowels)
```

Removing an item from a list is similar. To remove a specific item from the list, the `.remove(<item>)` function works well. If you want to remove just the last item from the list `.pop()` removes the last item and there is also a `del` keyword that deletes the item at the specified index, or even the entire list.

```python
menu = ['pizza', 'pasta', 'burger', 'pie', 'waffles', 'cake']
print(menu)
# let's remove 'pie', 'cake', and 'burger' off the menu list
menu.remove('pie')
print(menu)

menu.pop() # removing the last item -> 'cake'
print(menu)

del menu [2] # deleting item at index 2 -> 'burger'
print(menu)
```

Values in the list can also be swapped with another item as such:

```python
menu[1] = 'ice cream'
print(menu)
```

There are also many builtin functions such as `len(<list>)` which returns the length of the list, `.sort()` which gives back the list sorted.The sort function modifies the values in the list in place. 

```python
# built-in functions like len()
floats = [1.2, 2.3, 4.2, 4.5, 3.7]
print(floats)
floats.sort() # sorts the values in place
print(f'Sorted: {floats}')
length_of_list = len(floats)
print(f'List length: {length_of_list}')
```

Lists can also be sliced to a subset of that list. Let's slice the sorted floats list to only show the last 3 items.

**Tuples** are similar to lists, but the items in a tuple cannot be changed. Tuples can be slightly faster than lists, so using tuples can optimize your code in some instances. Now let's try a series of operations on a tuple.

### 1.4 Functions
We have already seen some builtin Python functions such as `len()`, `list()`, and `type()`, but Python also allows the space to create your own functions which perform a certain operation and return an output. To define a function, `def function_name():` must be at the top and all statements must be indented within the function itself. A function can also have parameters which specify arguments that the function can use. 

Let's create a function that converts a temperature in Fahrenheit to Celsius, taking a *temp* variable as the parameter input.

```python
# defining a function
def to_Celsius(temp):
    celsius = (temp - 32) * 5 / 9
    return celsius # the return statement outputs the value to wherever the function was called.
```

```python
# the data values in the fahrenheits list are the arbitrary average temperatures of each month of a year.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
fahrenheits = [28.4,32.0,37.5,50.2,59.4,68.1,82.7,96.4,75.2,49.3,31.4]
```

```python
celsius = [] # an empty list which will catch the celsius equivalents
for temperature in fahrenheits: # iterating over the list 
    # Here we call the function that converts the temperature to celsius and store the result in new_temp
    new_temp = to_Celsius(temperature)
    new_temp_rounded = round(new_temp, 2) # rounding to 2 decimal places
    celsius.append(new_temp_rounded) # adding the rounded result to the celsius list

print("Here are the temperatures in Celsuis:")
print(celsius)
```

Python **Generator Functions** are a simply way of creating iterators, and an extremely efficient approach when working with large data. A generator function returns an object which can be iterated over one value at a time.

A generator function contains:
- one or more `yield` statements (similar to `return` statement, but it pauses the function instead of completely terminating it.)
- Returns an iterator but doesn't start executing immediately at call.
- 'StopIteration' occurs automatically when the function terminates on future calls.

Generators are easy to implement, very efficient for memory, and great for representing infinite data streams.

Here's a diagram depicting the difference between Normal Python functions, and Generator function. 

<img src="../images/getting_started_images/01_normal-vs-generator.png">

(Source: https://codeburst.io/understanding-generators-in-es6-javascript-with-examples-6728834016d5)

Let's create a generator function to calculate the squares of numbers from a list, and iterate through the items by using `next()`.

```python
num_list = [2, 3, 4, 5, 6, 7, 8, 9, 10]

def square(numbers): # the generator function
    for num in numbers:
        yield num**2
```

```python
sq = square(num_list) # calling the function
next(sq) # first iteration
```

Now let's run this code until the function terminates. However, we know that a **StopIteration Traceback error** results from the function automatically terminating, and we want to avoid our code block resulting in an error.

Python's ***try except*** method is great for handling errors. The `try` condition checks the block of code for errors and the `except` block handles errors.
Below, the while loop which prints out the next squared number is placed within the try block to test for the error. The except block following directly catches the *StopIteration* error. 


```python
try:
    while sq: # all the next iterations
        print(next(sq))
except StopIteration:
    print("StopIteration error handled successfully")
```

Compared to normal functions, generator functions are much smaller in size. Let's create a the same function which returns the squares of a list of number in normal python way, and compare the sizes between the two types of functions using the `sys` module. 

```python
def normal_square(numbers):  # the normal squared function
    squared = []
    for num in numbers:
        squared.append(num**2)
    return squared
```

```python
normal_sq = normal_square(num_list)  # calling the function
normal_sq
```

```python
# size comparison normal function vs generator function
import sys

print(f"The size of the normal function is: {sys.getsizeof(normal_sq)} bytes.")
print(f"The size of the generator function is: {sys.getsizeof(sq)} bytes.")
```

## **2. Geopandas Basics**

Geopandas is a geographical wrapper around pandas, with abstracted focus on working with geospatial data (as its name suggests). Geopandas uses the *shapely* library to perform geometric operations and *matplotlib* for plotting maps.

### 2.1 Creating a polygon in a GeoDataFrame
Let's start by creating a geodataframe which will store the geometry of a polygon shape of the Location 'Canada'. 

```python
import geopandas as gpd
from shapely.geometry import Polygon

geodf = gpd.GeoDataFrame({"location": "Canada", "geometry": None}, index=[0])
geodf
```

Now that we have created the geodataframe, we must add in the polygon shape to the Geometry column. Let's add all longitude and latitude as tuple pairs, which are the boundary coordinate encompassing Canada. For the sake of simplicity, only 14 coordinates points are used.

```python
# Each tuple: (Long, Lat)
coordinates = [(-141.120889, 69.534518), (-141.120889, 59.888937), (-126.273476, 48.283193), (-95.338529, 48.283193), 
               (-83.968518, 45.920587), (-83.176974, 41.804078), (-67.904183, 46.679594), (-65.859763, 43.004647), 
               (-51.964095, 46.558860), (-60.282758, 66.618122), (-77.493945, 75.095633), (-57.783440, 82.494824), 
               (-78.891274, 83.480366), (-128.830872, 70.140364)]

poly = Polygon(coordinates) # create the polygon shape from the tuple list using the shapely package
poly
```

The output polygon looks a bit funky. Mainly because of the coordinate points chosen. Canada has a very irregular shape with many curvatures, 14 coordinate points cannot properly depict the shape of Canada. As an exercise, try and add more coordinates to get a more appropriate shape.

```python
geodf.loc[0, 'geometry'] = poly # let's add this polygon to the geometry column, in the Canada row. 
geodf
```

```python
print(geodf.crs)
```

We can see our geodataframe has not been assigned a Coordinate Reference System (CRS). This means our polygon exists in an arbitrary space. The CRS is important because it tells our Python program that the polygon is an area on the Earth. 
Let's assign our dataframe to EPSG:4326, which is the identifier for the World Geodetic System 1984 (WGS84), the standard CRS for Earth.

```python
geodf.crs = 4326
geodf.crs
```

We can save this geodataframe as a shapefile, and now it can imported as a georeferenced polygon into other softwares like ArcGIS and QGIS.

```python
geodf.to_file("data/canada.shp")
```

To actually read the data, `.read()` must be called, which will return a NumPy array depicting the raster.

*The parameter for read is 1, because our image contains a singular band. If an entire multispectral image is imported, then other integers can be passed to read a particular band of the image.*

```python
with rasterio.open(band2) as src: # src is the dataset reader
    ## Let's try and understand the image better.
    print(f'The raster is stored in a: {type(src.read(1))}')
    print(f'The data types: {src.read(1).dtype}')
    print(f'Image size is: {src.shape[0]} rows x {src.shape[1]} columns')
    print(f'The number of bands in the image: {src.count}')
    print(f'The projection: {src.crs}')
    print(f'The driver format: {src.driver}')
    print(f'The geographic transform of the image: \n{src.transform}\n')

    # We can most of this data directly from the image's metadata.
    print("Image metadata:")
    metadata = src.meta
    print(metadata) # shares the attributes of the image (the metadata)
```

If you look at the code block above, you will note multiple print statements nested in the `with` statement which opens band 2. These statements explain the raster band 2. The objects within the curly brackets `{}` are the methods of getting the information. For example, `src.crs` returns the CRS of the image, which is EPSG:32610. Most of this data can be accessed by calling the image's metadata. 
***Metadata*** is data describing the image data and it is important for preserving records of long-term datasets.

### 3.3 Performing the task of clipping the image

Each of the Landsat 8 images cover a large region on Earth, and sometimes we only want to focus on a specific area of interest (AOI). The imported image can directly be clipped to the desired AOI using rasterio. 
Here, we create a function which takes the path of the image, and returns a clipped image around the specified bounding area given by rasterio's windows function. This is an efficient way to view just the subset of the image because it does not require the entire image to be processed, proving another reason for why Cloud-optimized GeoTIFFs are an optimal format.

Let's get started by opening **band 5** of a new Landsat 8 image taken on May 10th, 2020 in Southern British Columbia (path 47, row 26).

```python
# NIR Spectral Band
band5 = 'https://storage.googleapis.com/gcp-public-data-landsat/LC08/01/047/026/LC08_L1TP_047026_20200510_20200526_01_T1/LC08_L1TP_047026_20200510_20200526_01_T1_B5.TIF'
```

```python
def get_clipped_img(url):
    with rasterio.open(url) as src:
        row_offset, col_offset = src.index(474013,5458927) # coordinate collected using QGIS software.
                                                              # width, height
        window = rasterio.windows.Window(col_offset, row_offset, 4000, 2000)
        # read only the section specified by the window
        clipped = src.read(1, window=window) 
        show(clipped, cmap='terrain') # using the 'terrain' colormap
    return clipped
```

```python
clipped_b5 = get_clipped_img(band5)
```

### 3.4 Performing calculation on the Raster Images
Doing computation on the raster data is an essential part of geospatial analysis, such as presenting the spectral indices of granules or even just normalizing the data for plotting purposed. Let's experiment by plotting a combination of Landsat bands to create an RGB True colour composite image. Let's begin by importing in the Red, Green, and Blue bands from the Landsat 8 image taken on *August 31st, 2020, in the Bay region south of California, US in Northwestern Mexico.*

Before we can plot the RGB image, the arrays must be normalized, only then can they be combined together. The array is normalized by dividing each array item by the total array maximum. (NumPy makes it quick to calculate with only 1 line, no for loop needed!) After that, using the NumPy `.dstack()` function to combine the three bands, we can plot the final RGB image.

```python
with rasterio.open(band2) as b2, rasterio.open(band3) as b3, rasterio.open(band4) as b4:
    # Read in the 3 bands
    blue = b2.read(1)
    green = b3.read(1)
    red = b4.read(1)
    
    # Normalize the numpy arrays
    redN = red/red.max()
    greenN = green/green.max()
    blueN = blue/blue.max()
    
    # Stack the 3 bands together (order matters)
    rgb = np.dstack((redN, greenN, blueN))

    # Using matplotlib, let's see how the True colour image looks
    plt.imshow(rgb)
    plt.show()
```

Great! After normalizing our data and stacking them together we now have an image of the Northern region in the Gulf of California looking like a picture taken from a satellite.

## **4. Xarray Basics**

Xarray is a Python package which makes it simple and efficient to work with *labelled multi-dimensional arrays*. It borrows the raw Numpy array structure with additional dimensions of labels, convenient for large datasets. Xarray provides the basic data structures used for EO analysis, as well as powerful tools for computation and visualization. Xarray distinguishes itself from many netCDF data tools as it provides data structures for in-memory analytics that both utilize and preserve dimensional labels. 

The two main Xarray data structures are:
- **DataArray** - comprised of a labelled, multi-dimensional array with:
    * `values`: a numpy N-Dimensional array with the array values.
    * `dims`: the dimensions for each axis in the array. (for example, ('x', 'y', 'z') or ('time', 'space'))
    * `coords`: a dictionary type of container of array labelling each of the points.
    * `attrs`: dictionary keeping the metadata of the array.
    
- **Dataset** - comprised of multiple variables that can share the same coordinates (similar to Pandas Dataframes), with the properties:
    * `data_vars`: dictionary-like container of DataArrays.
    * `dims`: dictionary mapping the dimension names to the length of the dimensions. (for example, {'x': 2, 'y': 4, 'z': 3})
    * `coords`:  a dictionary type of container of DataArrays labelling each of the points in data_vars.
    * `attrs`: dictionary keeping the metadata of the array.


### 4.1 Reading in Data as DataArray

Let's get started with a quick example to understand how Xarray works. We'll pass in band 2 from the raster image used in the Rasterio section. The band will be read in using Xarray's `.open_rasterio(<filename>)` function. After that, we will convert the numpy array into an Xarray DataArray of name 'blue' and plot the image using the native Xarray `.plot()` function.

```python
import xarray as xr
import numpy as np
```

```python
with xr.open_rasterio(band2) as src:
    blue = src.data
    blue_da = xr.DataArray(blue, name='blue')
    print(blue_da)
    blue_da.plot()
```

### 4.1 Merging DataArrays into a Dataset

With complex data, the extra-multidimensionality of Datasets is sometimes necessary. We have already plotted an RGB image using Rasterio, now let's try to perform the same task using Xarray functions by converting the bands from DataArrays together to one Dataset!

First let's initialize an empty dataset to which we will combine the DataArrays. Next, open all three bands (bands 2, 3, and 4) which make up Blue, Green, and Red, respectively. Similar to the Rasterio method, we must normalize the three bands so they can be stacked on the same scale. Then these NumPy arrays will be converted to DataArrays (as in the example above) before being merged to the Dataset. 


```python
xr_ds = xr.Dataset()

with xr.open_rasterio(band2) as src:
    blue = src.data
    blueN = blue/blue.max()
    xr_ds = xr_ds.merge(xr.DataArray(blueN, name='b2'))
    
with xr.open_rasterio(band3) as src:
    green = src.data
    greenN = green/green.max()
    xr_ds = xr_ds.merge(xr.DataArray(greenN, name='b3'))
    
with xr.open_rasterio(band4) as src:
    red = src.data
    redN = red/red.max()
    xr_ds = xr_ds.merge(xr.DataArray(redN, name='b4'))
```

Once the bands have been combined in one DataSet, we need to assign them as a separate Data Variable known as "rgb" which is the stacking of bands 4, 3, and 2 along their dimensions. 

```python
xr_ds['rgb'] = (("dim0", "dim1", "dim2"), np.vstack([xr_ds.b4.data,xr_ds.b3.data,xr_ds.b2.data]))
xr_ds
```

Now, we can plot the multi-band image using the "rgb" variable.

```python
xr_ds.rgb.plot.imshow()
```

As we can see from the Xarray displayed above, this dataset has six dimensions, three shared by the variables 'b2', 'b3', 'b4' (dim_0, dim_1, dim_2), and three for the combination of the bands for variable 'rgb' (dim0, dim1, dim2).

### 4.2 Types of Indexing
There are two main ways to index Xarray DataArrays and Datasets, label-based indexing and position-based indexing.

**Label-based indexing**

Label-based index allows to access the information of a coordinate by calling its label. This feature is inspired by pandas method of indexing.

```python
# Let's get the Dataset for one band from the layered RGB using dim0 to isolate 1 band.
xr_ds.sel(dim0=0)
```

**Position-based indexing**

Position-based index takes after the method of NumPy array indexing, with the power of labelled dimensions. 

```python
xr_ds.isel(dim_1=2, dim_2=5) 
```

### 4.3 Overlaying a Clipped Raster Image
Similar to how we used Rasterio's windows to select a subset of the image, let's create a numpy array which will contain the values for the clipped image. This array will be converted into a DataArray and merged with our existing Dataset. Let's read in a different band (band 1) of the same image to be clipped and overlayed the full RGB data. 

```python
with xr.open_rasterio(band1) as src:
    zero = np.zeros(src.data.shape)
    zero[:] = np.nan
    zero[:,3000:5000,2000:5000] = src.data[:,3000:5000,2000:5000]
    xr_ds = xr_ds.merge(xr.DataArray(zero, name='b1'))
xr_ds
```

```python
xr_ds.b1[0].plot.imshow() 
```

Now that we know how the clipped image and the full RGB image look in different plots, let's combine them together in one image. 

```python
fig, ax = plt.subplots(1,1, figsize=(15,15))
xr_ds.rgb.plot.imshow(ax=ax)
xr_ds.b1[0].plot.imshow(ax=ax, alpha=0.8)
```

The final merged dataset contains all four bands which plotted together look as the result above.
